### Combine columns (ccording to their similar content)


```python
# Define a custom function to combine columns
def combine_columns(row, columns):
    parts = []
    
    for column in columns:
        if not pd.isna(row[column]):
            parts.append(row[column])
    
    return ''.join(parts)
```

### Extract Origin Country


```python
# List of possible countries
countries_list = [
    'USA', 'UK', 'Canada', 'Mexico', 'Argentina', 'Brazil', 'Chile', 'Peru', 'Colombia',
    'United Kingdom', 'France', 'Germany', 'Italy', 'Spain', 'Russia', 'Netherlands', 'Switzerland',
    'Australia', 'New Zealand', 'Japan', 'China', 'India', 'South Korea', 'Singapore',
    'South Africa', 'Nigeria', 'Kenya', 'Egypt', 'Ghana', 'Morocco', 'Ethiopia', 'Tanzania',
    'Saudi Arabia', 'United Arab Emirates', 'Turkey', 'Iran', 'Israel', 'Qatar', 'Jordan',
    'Thailand', 'Indonesia', 'Malaysia', 'Philippines', 'Vietnam', 'Cambodia', 'Laos', 'Myanmar',
    'Africa', 'Antarctica', 'Asia', 'Europe', 'North America', 'Oceania', 'South America'
]

# Define a function to extract origin countries from the description
def extract_origin_countries(text):
    if isinstance(text, str):  # Check if the value is a string
        # List of keywords indicating country of origin
        keywords = ['origin', 'native', 'born', 'originally from', 'originating from', 'naturalizes', 'natural', 'habitats']

        # Create the country pattern dynamically
        country_pattern = r'(?:\b(?:' + '|'.join(countries_list) + r')\b)'

        origin_countries = []
        sentences = text.split('.')

        for sentence in sentences:
            if any(keyword in sentence.lower() for keyword in keywords):
                matches = re.findall(country_pattern, sentence, re.IGNORECASE)
                if matches:
                    origin_countries.extend(matches)

        return ', '.join(origin_countries) if origin_countries else np.nan
    else:
        return np.nan

# Apply the function to create the 'Origin Country' column
plants_df['Origin Country'] = plants_df['Introduction'] + plants_df['Description']
plants_df['Origin Country'] = plants_df['Origin Country'].apply(extract_origin_countries)


plants_df['Origin Country']
```
